```python
# Verrijking van de ICTU data met BAG gegevens.
matches = pd.merge(ictu, bag, 
                   how='left',
                   left_on=['Straat','Huisnummer', 'Huisletter', 'Toevoeging'],
                   right_on = ['_openbare_ruimte_naam_nummeraanduiding', 'huisnummer_nummeraanduiding',
                               'huisletter_nummeraanduiding', 'huisnummer_toevoeging_nummeraanduiding'])
```

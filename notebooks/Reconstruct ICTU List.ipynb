```python
# Selectie van zaken welke via ICTU/LAA zijn binnengekomen.
ICTU = zakenDataset.data[zakenDataset.data.beh_oms.str.contains('laa', case=False, regex=False)]
```

```python
# Zoek de bijbehorende stadia voor elke zaak in de ICTU lijst.
ICTU_zaak_ids = ICTU.zaak_id
ICTU_zaak_id_regex = '|'.join(ICTU_zaak_ids)
ICTU_stadia = stadiaDataset.data[stadiaDataset.data.zaak_id.str.match(ICTU_zaak_id_regex)]
```

```python
# Sanity check.
print(len(ICTU_zaak_ids))
print(len(ICTU_stadia.zaak_id.unique()))
check = len(ICTU_zaak_ids) == len(ICTU_stadia.zaak_id.unique())
print(f"Is het aantal ICTU zaak id's gelijk? (zou zo moeten zijn!): {check}")
```

```python
# Check of ik de stadia kan vinden die een indicatie zijn van meldingen uit de ICTU lijst.
# Dus: meldingen met alleen de sta_code's: H_GERVP_BI, H_GVP_CORP en H_JUR_PROC.
# Op dit moment lijken deze codes niet voor te komen.

# TODO: met Saban checken waarom dit het geval is.
# --> antwoord: Het zijn nieuwe codes die pas vanaf de week van 26 augustus worden ingevoerd in de database.
#               De ICT lijst is dus pas vanaf dit moment reconstrueren.
#
# Oplossing nu: andere sta_code's model laten staan voor H_GERVP_BI, H_GVP_CORP en H_JUR_PROC :)
# Namelijk: h_bd, h_ovd_vtbd en h_bdovhan.

groups = ICTU_stadia.groupby('zaak_id')
```

```python
# Deze groep willen we in elk geval terug vinden in onze ICTU_list!
groups.get_group('115892_1')
```

```python
# Construeer de ICTU list, door te kijken of de 3 codes voorkomen,
# en of er exact 3 codes zijn (wellicht een check ten overvloede, maar better safe than sorry).
ICTU_list = []
for zaak_id, group in groups:
    sta_codes = group.sta_code
    ICTU_codes_found = (set(sta_codes) == {'h_bd', 'h_ovd_vtbd', 'h_bdovhan'})
    ICTU_length_satisfied = len(sta_codes)
    if ICTU_codes_found and ICTU_length_satisfied:
        ICTU_list.append(zaak_id)
```
